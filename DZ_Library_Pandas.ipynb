### Задание 1
#### Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
df = pd.read_csv('ml-latest-small/ratings.csv')
f = pd.read_csv('ml-latest-small/movies.csv')
joined = df.merge(f, on='movieId', how='left')
film = joined[joined['rating'] == 5.0]
film.groupby('title').count().sort_values('rating', ascending=False)['rating'].idxmax()
```

### Задание 2
#### По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
filtered_countries = data[ (data['country']=='Estonia') | (data['country']=='Latvia') | (data['country']=='Lithuania') ]
year = filtered_countries[ (filtered_countries['year'] >= 2005 ) & (filtered_countries['year'] <= 2010 )]
category = year[ (year['category'] == 4 ) | (year['category'] == 12 ) | (year['category'] == 21 ) ]
quantity = category.query('quantity > 0')
pribaltica = quantity[['quantity']].sum()
pribaltica
```
