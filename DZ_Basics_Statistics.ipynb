### Задание 1. Загрузка данных Изучить представленный набор данных на основе описания его столбцов, загрузить его и оставить 8 столбцов для дальнейшего изучения: surgery, Age, rectal temperature, pulse, respiratory rate, temperature of extremities, pain, outcome.

```python
# Загрузка данных, замена нечитаемых символов на NaN
missing_values = ['?']
df = pd.read_csv('horse_data.csv', na_values = missing_values)
df.columns = ['surgery', 'Age', 'Hospital Number', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'peripheral pulse', 'mucous membranes', 'capillary refill time', 'pain', 'peristalsis', 'abdominal distension', 'nasogastric tube', 'nasogastric reflux', 'nasogastric reflux PH', 'rectal examination - feces', 'abdomen', 'packed cell volume', 'total protein', 'abdominocentesis appearance', 'abdomcentesis total protein', 'outcome', 'surgical lesion?', 'type of lesion1', 'type of lesion2', 'type of lesion3', 'cp_data']
df.head()
```

```python
# Удаление лишних столбцов
horse_data = df.drop(['Hospital Number', 'peripheral pulse', 'mucous membranes', 'capillary refill time', 'peristalsis', 'abdominal distension', 'nasogastric tube', 'nasogastric reflux', 'nasogastric reflux PH', 'rectal examination - feces', 'abdomen', 'packed cell volume', 'total protein', 'abdominocentesis appearance', 'abdomcentesis total protein', 'surgical lesion?', 'type of lesion1', 'type of lesion2', 'type of lesion3', 'cp_data'], axis=1)
horse_data
```

### Задание 2. Первичное изучение данных Проанализировать значения по столбцам, рассчитать базовые статистики, найти выбросы.

```python
# расчет базовых статистик
horse_data.describe()
```

#### Поиск выбросов по столбцам с количественным типом данных и не значительными колебаниями в целом

```python
q1_rec = horse_data['rectal temperature'].quantile(0.25)
q3_rec = horse_data['rectal temperature'].quantile(0.75)
iqr_rec = q3_rec - q1_rec
lower_bound_rec = q1_rec - (1.5 * iqr_rec) 
upper_bound_rec = q3_rec + (1.5 * iqr_rec)
remove_outliers_rec = horse_data[horse_data['rectal temperature'].between(lower_bound_rec, upper_bound_rec, inclusive=True)].sort_values('rectal temperature')
remove_outliers_rec
```

```python
# Выбросы
pd.concat([horse_data, remove_outliers_rec]).drop_duplicates(keep=False)
```

```python
q1_pul = horse_data['pulse'].quantile(0.25)
q3_pul = horse_data['pulse'].quantile(0.75)
iqr_pul = q3_pul - q1_pul
lower_bound_pul = q1_pul - (1.5 * iqr_pul) 
upper_bound_pul = q3_pul + (1.5 * iqr_pul)
remove_outliers_pul = horse_data[horse_data['pulse'].between(lower_bound_pul, upper_bound_pul, inclusive=True)].sort_values('pulse')
remove_outliers_pul
```

```python
# Выбросы
pd.concat([horse_data, remove_outliers_pul]).drop_duplicates(keep=False)
```

### Задание 3. Работа с пропусками Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
# Расчет количества пропусков
(horse_data.isna().mean() * 100).round(2)
```

```python
# Формирование нового дата фрейма без пропусков
hd = pd.read_csv('horse_data.csv', na_values = missing_values)
hd.columns = ['surgery', 'Age', 'Hospital Number', 'rectal temperature', 'pulse', 'respiratory rate', 'temperature of extremities', 'peripheral pulse', 'mucous membranes', 'capillary refill time', 'pain', 'peristalsis', 'abdominal distension', 'nasogastric tube', 'nasogastric reflux', 'nasogastric reflux PH', 'rectal examination - feces', 'abdomen', 'packed cell volume', 'total protein', 'abdominocentesis appearance', 'abdomcentesis total protein', 'outcome', 'surgical lesion?', 'type of lesion1', 'type of lesion2', 'type of lesion3', 'cp_data']
new_horse_data = hd.drop(['Hospital Number', 'peripheral pulse', 'mucous membranes', 'capillary refill time', 'peristalsis', 'abdominal distension', 'nasogastric tube', 'nasogastric reflux', 'nasogastric reflux PH', 'rectal examination - feces', 'abdomen', 'packed cell volume', 'total protein', 'abdominocentesis appearance', 'abdomcentesis total protein', 'surgical lesion?', 'type of lesion1', 'type of lesion2', 'type of lesion3', 'cp_data'], axis=1)
```

```python
#Пропуски столбца с болью заменяем медианой по столбцу,
#без взаимосвязи с другими показателями, т.к. определение боли субъективно
fill_median_by_pain = new_horse_data
fill_median_by_pain['pain'].fillna(new_horse_data['pain'].median(), inplace=True)
```

```python
#Пропуски столбца с ректальной температурой заменяем медианой по столбцу, 
#в соответствии с возможной взаимосвязью по наличию боли
fill_median_by_rectal_temperature = new_horse_data
fill_median_by_rectal_temperature['rectal temperature'].fillna(new_horse_data.groupby(['pain'])['rectal temperature'].transform('median'), inplace=True)
```

```python
#Пропуски столбца с хирургическим вмешательством заменяем медианой по столбцу,
#в соответствии с возможной взаимосвязью по ректальной температуре и наличию боли
fill_median_by_surgery = new_horse_data
fill_median_by_surgery['surgery'].fillna(new_horse_data.groupby(['rectal temperature', 'pain'])['surgery'].transform('median'), inplace=True)
```

```python
#Пропуски столбца с пульсом заменяем медианой по столбцу, 
#в соответствии с возможной взаимосвязью по ректальной температуре
fill_median_by_pulse = new_horse_data
fill_median_by_pulse['pulse'].fillna(new_horse_data.groupby('rectal temperature')['pulse'].transform('median'), inplace=True)
```

```python
#Пропуски столбца с частотой дыхания заменяем медианой по столбцу, 
#без взаимосвязи с другими показателями, т.к. полезность показателя частоты дыхания сомнительна из-за больших колебаний
fill_median_by_respiratory_rate = new_horse_data
fill_median_by_respiratory_rate['respiratory rate'].fillna(new_horse_data['respiratory rate'].median(), inplace=True)
```

```python
#Пропуски столбца с температурой конечностей заменяем медианой по столбцу,
#в соответствии с возможной взаимосвязью по ректальной температуре
fill_median_by_temperature_of_extremities = new_horse_data
fill_median_by_temperature_of_extremities['temperature of extremities'].fillna(new_horse_data.groupby('rectal temperature')['temperature of extremities'].transform('median'), inplace=True)
```

```python
#Пропуски столбца с состоянием лошади заменяем медианой по столбцу,
#в соответствии с возможной взаимосвязью по ректальной температуре, проведению операции и возрастом
fill_median_by_outcome = new_horse_data
fill_median_by_outcome['outcome'].fillna(new_horse_data.groupby(['rectal temperature', 'surgery', 'Age'])['outcome'].transform('median'), inplace=True)
```
