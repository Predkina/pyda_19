## Домашнее задание к лекции «Функции»

#### Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

#### Исходные данные имеют следующую структуру:

##### 1. перечень всех документов

#### documents = [{‘type’: ‘passport’, ‘number’: ‘2207 876234’, ‘name’: ‘Василий Гупкин’},{‘type’: ‘invoice’, ‘number’: ‘11-2’, ‘name’: ‘Геннадий Покемонов’},{‘type’: ‘insurance’, ‘number’: ‘10006’, ‘name’: ‘Аристарх Павлов’}]

##### 2.перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)

#### directories = {‘1’: [‘2207 876234’, ‘11-2’],‘2’: [‘10006’],‘3’: []}

#### Общие требования к программе:

#### код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется); в коде отсутствуют глобальные переменные (за исключением documents и directories); пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды “q”).

```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
```

```python
# Пункт 1. Пользователь по команде “p” может узнать владельца документа по его номеру;
def people_document(arg_number):
    for doc_number in documents:
        if doc_number["number"] == arg_number:
            print(' Владелец документа: ', doc_number["name"])
            break
    else:
        print('Документ не найден в базе')
        
        
# Пункт 2. Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится;
def shelf_number(arg_number):
    shelf_break = False
    for shelf_directory in directories.items():
        for doc_number in shelf_directory[1]:
            if doc_number == arg_number:
                print('Документ хранится на полке:', shelf_directory[0])
                shelf_break = True
                break
        if shelf_break == True:
            break
    else:
        print('Документ не найден в базе')


# Пункт 3. Пользователь по команде “l” может увидеть полную информацию по всем документам в виде - №: 2207 876234, тип: passport, владелец: Василий Гупкин, полка хранения: 1;
def list_documents():
     for person in documents:
            z = list(filter(lambda e: person['number'] in e[1], directories.items()))[0]
            person['directory'] = z[0]
            print('№: ' + person['number'] + ',', 
              'тип: ' + person['type'] + ',', 
              'владелец: ' + person['name'] + ',', 
              'полка хранения: ' + person['directory'])
        
# Пункт 4. Пользователь по команде “as” может добавить новую полку;
def add_shelf_num(arg_new_shelf):
    if arg_new_shelf in directories.keys():
        print('Такая полка уже существует. Текущий перечень полок:', ','.join(directories.keys()))
    else:
        new_dict = {arg_new_shelf: []}
        directories.update(new_dict)
        print('Полка добавлена. Текущий перечень полок:', ','.join(directories.keys()))
        
# Пункт 5. Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая);
def del_shelf(number):  
    if number in directories and directories[number]:
        print ('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ' + ', '.join(directories.keys()))
    else:
            if number in directories and not directories[number]:
                del directories[number]
                print('Полка удалена. Текущий перечень полок: ' + ', '.join(directories.keys()))
            else:
                    print('Такой полки не существует. Текущий перечень полок:' + ', '.join(directories.keys()))
                    
# Основное меню для запуска команд
def main_menu():
    while True:
        command = input('\n \
    Выберете одну из команд: p, s, l, as, ds \n \
    Для выхода наберите: q \n \
    Введите команду: ')
        if command == 'p':
            people_document(input('Введите номер документа: '))
        elif command == 's':
            shelf_number(input('Введите номер документа: '))
        elif command == 'l':
            list_documents()
        elif command == 'as':
            add_shelf_num(input('Введите номер полки: '))
        elif command == 'ds':
            del_shelf(input('Введите номер полки: '))
        elif command == 'q':
            break
        else:
            print('Вы ввели команду не корректно, повторите ввод.')


main_menu()
```
